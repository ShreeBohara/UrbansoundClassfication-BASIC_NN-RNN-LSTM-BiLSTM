```python
prec_dos = []
prec_normal = []
prec_probe = []
prec_r2l = []
prec_u2r = []
prec_dos2 = []
prec_normal2 = []
prec_probe2 = []
prec_r2l2 = []
prec_u2r2 = []
models=[]

for i in performance_data:
    prec_dos.append(i['precision'][0])
    prec_normal.append(i['precision'][1])
    prec_probe.append(i['precision'][2])
    prec_r2l.append(i['precision'][3])
    prec_u2r.append(i['precision'][4])
    prec_dos2.append(i['precision'][0])
    prec_normal2.append(i['precision'][1])
    prec_probe2.append(i['precision'][2])
    prec_r2l2.append(i['precision'][3])
    prec_u2r2.append(i['precision'][4])
    models.append(i['model'])
print(len(prec_dos))
```

# Accuracy

```python
accuracy = []
models = []

for i in performance_data:
    accuracy.append(i['accuracy'])
    models.append(i['model'])
accuracy
```

```python
plt.figure(figsize=(25,15))

br1 = np.arange(len(prec_dos))
br1 = br1*5

plt.bar(x=br1 , height=accuracy)

plt.xlabel('MODELS', fontweight ='bold', fontsize = 35)
plt.ylabel('ACCURACY', fontweight ='bold', fontsize = 35)
plt.xticks(br1,models)
plt.show();
```
